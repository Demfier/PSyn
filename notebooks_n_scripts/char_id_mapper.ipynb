```python
import pandas as pd
from PSyn import data_operators
import os
import json
```

```python
source_csv_path = 'data/task1/output/source_csvs/'
map_store_path = 'data/task1/output/char_id_map/'
source_files = os.listdir(source_csv_path)
```

```python
for file in source_files:
    if (map_store_path + file) in os.listdir(map_store_path):
        continue
    source_csv = pd.read_csv(source_csv_path + file)
    print(file)
    chars = data_operators.extract_alphabets(source_csv['source'].dropna())
    char_id_map = {}
    for i, char in enumerate(chars):
        char_id_map[char] = i
    with open(map_store_path + file, 'w') as j:
        json.dump(char_id_map, j)
```
