```python
from PSyn import brain
import pandas as pd
import numpy as np
import time
import os
```

```python
adj_mat_root = 'data/task1/output/bigram_dfs/normalized/'
opn_df_root = 'data/task1/output/opn_dfs/opn_count/'
languages = os.listdir(opn_df_root)
```

```python
accuracy_det = dict()
for lang in languages:
    try:
        accuracy_det[lang] = brain.decision_tree(adj_mat_root + lang.replace('.csv', '.p'), opn_df_root + lang)
    except ValueError:
        print('ValueError. Probably the lack of features to train')
        continue
    except:
        print('Dont have the slightest of idea what is!')
```

```python
df = pd.DataFrame.from_dict(accuracy_det).transpose()
```

```python
df.to_csv('data/task1/output/prediction/decision_tree/accuracy.csv')
```
