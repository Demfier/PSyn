```python
import pandas as pd
import numpy as np
from PSyn import matrix_functions
import os
import operator
import json
```

```python
trans_mat_path = 'matrices/'
trans_matrices = os.listdir(trans_mat_path)
normalized = map(lambda x: 'normalized_' + x, trans_matrices)
for mat in trans_matrices:
    if mat not in normalized:
        continue
    matrix = pd.read_pickle(trans_mat_path + mat)
    node_rank = matrix_functions.random_walk(matrix)
    print(node_rank[:10])

hindi = pd.read_pickle('normalized_hindi.p')
node_rank = matrix_functions.random_walk(hindi)
language_wise_node_rank = {'hindi': node_rank}
with open('language_wise_node_rank.json', 'w') as f:
    json.dump(language_wise_node_rank, f)
with open('language_wise_node_rank.json', 'r') as f:
    data = json.load(f)
sorted(data['hindi'], key=lambda x: x[2])
```
